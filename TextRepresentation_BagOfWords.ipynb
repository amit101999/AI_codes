```python
df = pd.read_csv('spam.csv')
df.head()
```

```python
df.Category.value_counts()
```

```python
# creating a new column with numerical values
df['Spam'] = df['Category'].apply(lambda x : 1 if x == 'spam' else 0)
df.head()
```
